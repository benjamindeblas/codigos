## Homologacion de tipos de datos


```python
df.columns = ['trip_id','taxi_id','trip_st_ts','trip_end_ts','trip_seconds','trip_miles','fare','tips','tolls', 'extras','trip_total'
,'payment_type','company','pu_lat','pu_lon','do_lat','do_lon',]
```

```python
df.head()
```

```python
for c in ['trip_st_ts','trip_end_ts']:
 df[c]= pd.to_datetime(df[c], format= '%m/%d/%Y %I:%M:%S %p')
```

```python
df.columns
```

```python
for c in ['trip_seconds','trip_miles','fare','tips','tolls'	,'extras','trip_total','pu_lat','pu_lon',	'do_lat','do_lon']:
    df[c]= pd.to_numeric(df[c],errors='coerce')
```

## Agrupamiento


```python
aux = df.loc[df['payment_type'].isin(['Credit Card','Mobile'])].reset_index(drop=True)
aux
```

```python
aux = df.loc[df['payment_type'].isin(['Credit Card','Mobile'])].reset_index(drop=True)
aux = aux[['company','tips','trip_total']].groupby('company').sum()
aux['%propina']= aux['tips']/aux['trip_total']
aux
```

```python
aux = df.loc[df['payment_type'].isin(['Credit Card','Mobile'])].reset_index(drop=True)
aux[['payment_type','tips','trip_total']].groupby('payment_type').agg(['mean','count'])
```

```python
aux = df[['payment_type','tips','trip_total']].groupby('payment_type').sum()
aux['%propina']= aux['tips']/aux['trip_total']
aux
```
