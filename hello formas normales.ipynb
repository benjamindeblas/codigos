```python
query = """
SELECT 
    nombre_proyecto, SUM(costo) AS costo_total_proyecto
FROM
    tbl_asignacion A
        INNER JOIN
    tbl_proyecto B ON A.num_proyecto = B.num_proyecto
        INNER JOIN
    tbl_empleado C ON A.num_empleado = C.num_empleado
        INNER JOIN
    tbl_puesto D ON C.puesto = D.puesto
GROUP BY nombre_proyecto
ORDER BY costo_total_proyecto DESC
;
"""
```
